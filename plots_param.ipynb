### Using priors from GWTC 1 paper. alpha, beta = [-4, 12]; mmin = [5, 10], mmax = [30, 100]. Nsamples= 10000


```python
import emcee
filename = "data/gwtc3_5000_trial.txt"
backend = emcee.backends.HDFBackend(filename) 
flat_samples = backend.get_chain(discard=10, thin=100, flat=True)
```

```python
import corner

labels = ['alpha', 'beta', 'mmin', 'mmax', 'lamb', 'mu', 'sigma']
# labels = ['alpha', 'beta', 'mmin', 'mmax']


fig = corner.corner(
    flat_samples, labels=labels, smooth=1.2
);
```
