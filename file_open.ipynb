```python
events, log_likelihood = [], []
```

```python
import csv
```

```python
with open('ext.txt', 'r') as f:
    for l in f:
        blocks = l.strip().split(" ")
        events.append(blocks[0])
        log_likelihood.append(blocks[-1])
```

```python
event100, event200 = [], []
```

```python
logL1, logL2 = [], []
```

```python
i = 0
for event in events:
    line = event.split("_")
    if float(line[-1]) == 100:
        event100.append(line[0]+line[1])
        logL1.append(float(log_likelihood[i]))
    else:
        event200.append(line[0]+line[1])
        logL2.append(float(log_likelihood[i]))
    i+=1
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.xlabel('events')
plt.ylabel('likelihood')
plt.plot(event100, np.exp(logL1), '-bo') # blue curve = 100 epochs
plt.plot(event200, np.exp(logL2), '-ro') # red curve = 200 epochs
```

```python
events, log_likelihood = [], []
with open('ext1.txt', 'r') as f:
    for l in f:
        blocks = l.strip().split(" ")
        events.append(blocks[0])
        log_likelihood.append(blocks[-1])
```
